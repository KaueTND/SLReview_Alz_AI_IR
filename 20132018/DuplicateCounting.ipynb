```python
import pandas as pd
import numpy as np
import csv
```

CRIAÇÃO DAS BASES DE DADOS, E TRADUÇÃO DELAS PARA A BIBLIOTECA PANDAS

```python
springer = pd.read_csv('SPRINGERporCSV.csv')
springer_df = springer[['Item Title','Publication Year','Item DOI']].copy()
springer_df.columns = ['Title','Year','DOI']
springer_df['Base'] = 'SPRINGER'
```

```python
acm = pd.read_csv('ACMporCSV.csv')
acm_df = acm[['title','year','doi']].copy()
acm_df.columns = ['Title','Year','DOI']
acm_df['Base'] = 'ACM'
```

```python
ieee = pd.read_csv('IEEEporCSV.csv',sep=',',encoding = "ISO-8859-1")
ieee_df = ieee[ieee.columns[[0,5,13]]].copy()
ieee_df.columns = ['Title','Year','DOI']
ieee_df['Base'] = 'IEEE'
```

```python
scopus = pd.read_csv('SCOPUSporCSV.csv',sep=',',encoding = "ISO-8859-1")
scopus_df = scopus[scopus.columns[0:3]].copy()
scopus_df['Base'] = 'SCOPUS'
```

```python
def get_DOI(description):
    splitted = description.split('doi: ')
    if(len(splitted)>1):
        return splitted[1].split(' ')[0][:-1]
    else:
        return np.nan

pubmed = pd.read_csv('PUBMEDporCSV.csv',sep=',',encoding = "ISO-8859-1",index_col=False)    
pubmed['DOI'] = pubmed['Details'].apply(get_DOI) 
pubmed['Year'] = pubmed['ShortDetails'].apply(lambda x:x.split('.')[1])
pubmed_df = pubmed[['Title','Year','DOI']].copy()
pubmed_df['Base'] = 'PUBMED'
```

```python
webofscience = pd.read_csv('savedrecs(2).txt',sep='\t',encoding = "ISO-8859-1",quoting=csv.QUOTE_NONE)
webofscience_df = webofscience[['CA','PD','AR']].copy()
webofscience_df.columns = ['Title','Year','DOI']
webofscience_df['Base'] = 'WOS'
webofscience_df
```

GERAÇÃO DE UMA BASE QUE CONTENHA TODOS OS REGISTROS E OS COM DOI NULL

```python
#Juntando as bases de dados
all_bases_df = pd.concat((springer_df, acm_df, ieee_df, webofscience_df, scopus_df, pubmed_df)) 
```

```python
#Quantos registros existem por base de dados
all_bases_df['Base'].value_counts()
```

```python
((all_bases_df['DOI'].str.lower()).value_counts()).count()
```

```python
doi = all_bases_df['DOI'].value_counts()
novodoi = pd.DataFrame(all_bases_df['DOI'].value_counts())
array = doi.index.get_values()
novodoi['index'] = array
doimaiorq2 = np.array(novodoi[novodoi['DOI']>1]['index'])

all_bases_df[all_bases_df['DOI'].isin(doimaiorq2)].sort_values(by='DOI',ascending=True)

#doi['index'] = array
#doi
```

```python
#Após a remoção de artigos duplicados
data_after_duplicated_remove = all_bases_df.drop_duplicates(subset='DOI',keep='last')
data_after_duplicated_remove['Base'].value_counts()
```

```python
#Agrupando Bases por DOI
new_test_base = all_bases_df[['DOI','Base']].copy()
grouped_base_by_DOI = new_test_base.groupby('DOI')['Base'].apply(list)
#grouped_base_by_DOI['DOI'=='10.1001/jamaneurol.2013.1321']
grouped_base_by_DOI = pd.DataFrame(grouped_base_by_DOI)
grouped_base_by_DOI.reset_index(inplace=True)
```

```python
#Merge das bases de dados criadas

result = pd.merge(data_after_duplicated_remove.sort_values(by='Base').drop('Base',axis=1), grouped_base_by_DOI, how='right', on=['DOI'])
```

```python
doi_null_base = all_bases_df[all_bases_df['DOI'].isnull()].sort_values(by='Title',ascending=True)
```

EXPORTAR PARA DOCUMENTO EXCEL

```python
writer = pd.ExcelWriter('SystematicReviewAlzheimer.xls')
result.to_excel(writer,'After Remove Duplication')
doi_null_base.to_excel(writer,'DOI Null')
writer.save()
```
